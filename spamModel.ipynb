```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import joblib

# Step 1: Load dataset
df = pd.read_csv(r"C:\Users\LAPTOP INSIDE\allSmallLettersSpam1.csv", usecols=['label', 'message'])

# Step 2: Split features and labels
X = df['message']
y = df['label']

# Step 3: Train-test split (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Step 4: TF-IDF Vectorization
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Step 5: Train Multinomial Naive Bayes Model
nb_model = MultinomialNB()
nb_model.fit(X_train_vec, y_train)

# Step 6: Predict
y_pred = nb_model.predict(X_test_vec)

# Step 7: Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f"\nðŸŽ¯ Naive Bayes Accuracy (70/30 split): {accuracy:.2%}\n")

print("ðŸ“Š Classification Report:")
print(classification_report(y_test, y_pred))

# Step 8: Confusion Matrix
cm = confusion_matrix(y_test, y_pred, labels=['ham', 'spam'])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['ham', 'spam'])

plt.figure(figsize=(6, 6))
disp.plot(cmap=plt.cm.Blues)
plt.title("ðŸ“ˆ Confusion Matrix - Naive Bayes Spam Classifier")
plt.show()

# Step 9: Save the model and vectorizer
joblib.dump(nb_model, r"C:\Users\LAPTOP INSIDE\naivebayes_spam_model.pkl")
joblib.dump(vectorizer, r"C:\Users\LAPTOP INSIDE\naivebayes_vectorizer.pkl")

print("\nâœ… Naive Bayes model and vectorizer saved successfully.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import joblib

# Step 1: Load dataset
df = pd.read_csv(r"C:\Users\LAPTOP INSIDE\allSmallLettersSpam1.csv", usecols=['label', 'message'])

# Step 2: Features and labels
X = df['message']
y = df['label']

# Step 3: Train-test split (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

# Step 4: TF-IDF Vectorization
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Step 5: Train Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_vec, y_train)

# Step 6: Predictions
y_pred = rf_model.predict(X_test_vec)

# Step 7: Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f"\nðŸŽ¯ Random Forest Accuracy (70/30 split): {accuracy:.2%}\n")

print("ðŸ“Š Classification Report:")
print(classification_report(y_test, y_pred))

# Step 8: Confusion Matrix
cm = confusion_matrix(y_test, y_pred, labels=['ham', 'spam'])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['ham', 'spam'])

plt.figure(figsize=(6, 6))
disp.plot(cmap=plt.cm.Greens)
plt.title("ðŸ“ˆ Confusion Matrix - Random Forest Spam Classifier")
plt.show()

# Step 9: Save model and vectorizer
joblib.dump(rf_model, r"C:\Users\LAPTOP INSIDE\rf_spam_model.pkl")
joblib.dump(vectorizer, r"C:\Users\LAPTOP INSIDE\rf_vectorizer.pkl")

print("\nâœ… Random Forest model and vectorizer saved successfully.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Step 1: Load dataset
df = pd.read_csv(r"C:\Users\LAPTOP INSIDE\allSmallLettersSpam1.csv", usecols=['label', 'message'])

# Step 2: Split into features/labels
X = df['message']
y = df['label']

# Step 3: Train-test split (70/30)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Step 4: TF-IDF Vectorizer
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Step 5: Initialize models
models = {
    "Naive Bayes": MultinomialNB(),
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42)
}

# Step 6: Train and evaluate
accuracies = {}
for name, model in models.items():
    model.fit(X_train_vec, y_train)
    y_pred = model.predict(X_test_vec)
    acc = accuracy_score(y_test, y_pred)
    accuracies[name] = acc
    print(f"{name} Accuracy: {acc:.2%}")

# Step 7: Plot comparison
plt.figure(figsize=(8, 5))
plt.bar(accuracies.keys(), accuracies.values(), color=['blue', 'orange', 'green'])
plt.ylabel('Accuracy')
plt.ylim(0.8, 1)
plt.title('ðŸ“ˆ Accuracy Comparison of Spam Detection Models')
plt.grid(axis='y')
plt.show()
print('\n\n\nit is cleare that the accuracy of Naive bayes algo is best\nobviously i will select this for my model')
```
